Use `nx.watts_strogatz_graph` to generate 3 graphs with 500 nodes each, average degree = 4, and rewiring probablity $p = 0, 0.1,$ and  $1$. Calculate the average shortest path length $\langle d \rangle$ for each one. Describe what happens to the network when $p = 1$.



```python
# Create a function that calculate the average shortest path <d> of a graph

def avg_shortest_path(G):
    # Get all pairs of nodes
    pairs = nx.all_pairs_shortest_path_length(G)
    # Sum all the shortest paths
    sum = 0
    for node, paths in pairs:
        for path in paths:
            sum += paths[path]
    # Divide by the number of pairs
    n = len(G.nodes())
    return sum/(n*(n-1))

# Create multiple WS graphs with 500 nodes, average degree 4 and p = 0, p=0.1, p=1
G1 = nx.watts_strogatz_graph(500, 4, 0)
G2 = nx.watts_strogatz_graph(500, 4, 0.1)
G3 = nx.watts_strogatz_graph(500, 4, 0.999999)

# Calculate the average shortest path for each graph
avg1 = avg_shortest_path(G1)
avg2 = avg_shortest_path(G2)
avg3 = avg_shortest_path(G3)

# Print the results
print("Average shortest path for p=0: " + str(avg1))
print("Average shortest path for p=0.1: " + str(avg2))
print("Average shortest path for p=0.999999: " + str(avg3))

random_graph_avg = np.log(500)/np.log(4)
print("Random graph average: " + str(random_graph_avg))
```

> * First create a graph consisting of a single link. (You can call the nodes anything, but I would simply use integers as names).

```python
# Create a graph with a single link
G = nx.Graph()
G.add_nodes_from([1,2])
G.add_edge(1,2)
```

> * Now add another node, connecting one of the existing nodes in proportion to their degree.

```python
# Function adding a new node to a graph and connecting it to node proportional to their degree
def add_node_and_link_to_hub(G):
    # Get the degree of each node
    degrees = G.degree()
    # Get the sum of all degrees
    sum = 0
    for (node, degree) in degrees:
        sum += degree
    # Get the probability of each node
    probabilities = []
    for (node, degree) in degrees:
        probabilities.append(degree/sum)
    # Get the node to connect to
    node_to_connect_to = np.random.choice(range(1, len(G.nodes)+1), p=probabilities)
    # Add the new node
    G.add_node(len(G.nodes())+1)
    # Connect the new node to the chosen node
    G.add_edge(len(G.nodes()), node_to_connect_to)
    return G

# Add a new node
G = add_node_and_link_to_hub(G)

# Print the network
nx.draw(G, with_labels=True)

```

> Next step is to explore the [Friendship paradox](https://en.wikipedia.org/wiki/Friendship_paradox). This paradox states that _almost everyone_ has fewer friends than their friends have, on average\*. This sounds crazy, but is actually an almost trivial consequence of living in a social network with a power-law degree distribution. The explanation is that almost everyone is friends with a hub, that drives up the average degree of the friends. Let's explore that in the 5000 node BA network we've just generated. Do the following:

> * Pick a node $i$ at random (e.g. use `random.choice`). [Find its degree](https://networkx.github.io/documentation/stable/reference/classes/generated/networkx.Graph.degree.html).

> * Find $i$'s [neighbors](https://networkx.github.io/documentation/networkx-1.10/reference/generated/networkx.Graph.neighbors.html?highlight=neighbors#networkx.Graph.neighbors). And calculate their average degree.

> * Compare the two numbers to check if it's true that $i$'s friends (on average) have more friends than $i$.

```python
def check(G):
    random_node = rd.choice(list(G.nodes()))

    # Get the neighbors of the random node
    neighbors = [n for n in G.neighbors(random_node)]

    # Calculate the average degree of the neighbors
    sum = 0
    for neighbor in neighbors:
        sum += G.degree(neighbor)

    if(len(neighbors) == 0):
        return check(G)
    avg_degree = sum/len(neighbors)

    if(G.degree(random_node) > avg_degree):
        return 0
    else:
        return 1
```

> * Do this 1000 times. How many out of those 1000 times is the friendship paradox true?

```python
def check_1000(G):
    nb_iter = 1000
    total = 0
    for i in range(nb_iter):
        total += check(G)

    print(f"It is the case {total} times over {nb_iter} iterations.")
check_1000(G)
```
